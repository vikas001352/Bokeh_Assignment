```python
ANS-1

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure object
p = figure(title='Simple Scatter Plot', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add scatter glyphs to the figure
p.scatter(x, y, size=10, color='blue', alpha=0.6)

# Show the plot
output_notebook()  # This line is optional and specific to Jupyter notebooks.
show(p)


ANS-2

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure object
p = figure(title='Scatter Plot with Circles', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add scatter glyphs to the figure
p.circle(x, y, size=10, color='blue', alpha=0.6)

# Show the plot
output_notebook()  # This line is optional and specific to Jupyter notebooks.
show(p)



In this example, we use Bokeh's figure() function to create a plot figure. The circle() method is then used to add scatter glyphs to the figure. The circle() method takes the following arguments:

x: The x-coordinates of the data points.
y: The y-coordinates of the data points.
size: The size of the circles in screen units.
color: The color of the circles.
alpha: The transparency of the circles (0 for completely transparent, 1 for fully opaque).
The x and y lists represent the data points for the scatter plot. We then specify the size, color, and alpha values for the circles. The resulting plot will display the scatter plot with blue circles at the specified data points.

Bokeh offers many other types of glyphs and customization options, allowing you to create a wide range of visualizations. For example, you can use line(), bar(), triangle(), or other glyph methods to create different types of plots. Additionally, you can customize the plot further by adjusting axis labels, legends, and plot styling.


ANS-3


Customize the Title:
To set the title of the plot, you can use the title attribute of the figure object. You can also customize the title font size, font style, and alignment.


from bokeh.plotting import figure, show

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure object with a custom title
p = figure(title='Customized Title', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add scatter glyphs to the figure
p.circle(x, y, size=10, color='blue', alpha=0.6)

# Show the plot
show(p)


Customize the Axes:
You can customize the axes by setting their labels, font size, font style, and other properties

from bokeh.plotting import figure, show

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure object with custom axis labels
p = figure(title='Scatter Plot with Custom Axes', x_axis_label='Custom X-axis', y_axis_label='Custom Y-axis')

# Add scatter glyphs to the figure
p.circle(x, y, size=10, color='blue', alpha=0.6)

# Customize the axis properties
p.xaxis.axis_label_text_font_size = '14pt'
p.yaxis.axis_label_text_font_style = 'italic'

# Show the plot
show(p)


Customize the Legend:
You can add a legend to the plot to label different data series or categories.


from bokeh.plotting import figure, show

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [6, 7, 2, 4, 5]
y2 = [3, 2, 1, 3, 2]

# Create a figure object
p = figure(title='Scatter Plot with Legend', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add scatter glyphs to the figure with custom legend labels
p.circle(x, y1, size=10, color='blue', alpha=0.6, legend_label='Series 1')
p.square(x, y2, size=10, color='green', alpha=0.6, legend_label='Series 2')

# Add a legend to the plot
p.legend.title = 'Legend Title'
p.legend.location = 'top_left'
p.legend.label_text_font_size = '12pt'

# Show the plot
show(p)


ANS-4


from bokeh.io import curdoc
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.layouts import column
import numpy as np

# Create initial data
x = np.random.random(100)
y = np.random.random(100)

# Create ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y))

# Create a scatter plot
plot = figure(title='Real-Time Scatter Plot', width=600, height=400)
plot.circle('x', 'y', source=source, size=10, color='blue', alpha=0.6)

# Function to update the plot data
def update_data():
    new_data = dict(x=np.random.random(100), y=np.random.random(100))
    source.data = new_data

# Add a button to trigger data update
from bokeh.models import Button
button = Button(label="Update Data")
button.on_click(update_data)

# Arrange the plot and button in a layout
layout = column(plot, button)

# Add the layout to the current document
curdoc().add_root(layout)


ANS-5

```
